```python
import xarray as xr
import numpy as np
import pandas as pd
import os
```

```python
PATH_RUN = "/net/pc200256/nobackup/users/most/output/ERA5_2023_PD/"
ds_demand = xr.open_dataset(PATH_RUN + "energy_demand/era5_demand_d_1950_2022.nc")
da_demand = ds_demand['demand']
da_ror = xr.open_dataarray(PATH_RUN + "agg_production/energy_prod_ror/era5_ror0.75_d_1950_2022_30min_agg.nc")
ds_wind_solar = xr.open_dataset(PATH_RUN + "agg_production/energy_prod_wind_solar/era5_prod_d_1950_2022_agg.nc")
ds_aggregated = xr.merge([da_demand, da_ror, ds_wind_solar])
```
