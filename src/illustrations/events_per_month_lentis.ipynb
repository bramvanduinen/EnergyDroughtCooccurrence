```python
import xarray as xr
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
DROUGHTS_FOLDER = "/usr/people/duinen/MSc-thesis/src/find_energydroughts/data"
DROUGTHS_FILE = f"LENTIS_FULLYEAR_v2_netto_demand_el7_full_LENTIS_2023_PD_1600_events.csv"
ed_df = pd.read_csv(os.path.join(DROUGHTS_FOLDER, DROUGTHS_FILE), index_col=0)

ed_df['start_time'] = pd.to_datetime(ed_df['start_time'])
ed_df['end_time'] = pd.to_datetime(ed_df['end_time'])
event_month = ed_df['start_time'].dt.month

plt.figure(figsize = (10, 6), dpi = 300)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.hist(event_month, bins=np.arange(0.5, 13.5, 1), rwidth=0.8)
plt.xticks(np.arange(1, 13), months, fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Event month', size = 14)
plt.ylabel('Number of events', size = 14)
```

```python
event_month_counts = event_month.value_counts()
selected_months_counts = event_month_counts.loc[[11, 12, 1, 2, 3]].sum()
print("The percentage of events in November-March season is:", (selected_months_counts / event_month_counts.sum()) * 100)
```
