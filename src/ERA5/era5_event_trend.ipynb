```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import utils as ut
from config_ERA5 import (
    COUNTRIES,
    PATH_CLUSTERS,
    PATH_DATA,
    PATH_ED_REGION,
    REGIONS,
    VERSION,
)
```

```python
df_events = pd.read_csv(os.path.join(PATH_ED_REGION + 'max_drought_regions_netto_demand_el7_winter_ERA5_2023_PD_noHydro_73_events_with_quantity.csv'), index_col=0)
```

```python
regions = ['N', 'B', 'NW', 'C', 'E', 'IB']
region_names = ['Northern', 'Baltic', 'Northwestern', 'Central', 'Eastern', 'Iberia']
alphabet = 'abcdef'
fontsize = 14

fig, axs = plt.subplots(3, 2, figsize=(15, 10), sharex=True, sharey = True, dpi = 300)
for i, region in enumerate(regions):
    ax = axs[i // 2, i % 2] 
    years_region = pd.to_datetime(df_events.query('country == @region')['start_time']).dt.year
    ax.hist(years_region, bins = 30)
    if i // 2 == 2:  # Bottom row
        ax.set_xlabel('Year', fontsize = fontsize)
    if i % 2 == 0:  # Left column
        ax.set_ylabel('Number of events', fontsize = fontsize)
    ax.set_title(f"({alphabet[i]}) {region_names[i]}", loc = "left", fontsize = fontsize)
    ax.tick_params(labelsize = fontsize)
plt.tight_layout()
```
